# Реализация модели

## Имплементация (6 баллов)

```python
from sklearn.base import BaseEstimator, ClassifierMixin

class MySGDClassifier(BaseEstimator, ClassifierMixin):
    """
    Предполагается, что в выборке всегда 2 класса
    """
    
    def __init__(self, C=1, alpha=0.01, max_epoch=10, random_state=None):
        """
        C - коэф. регуляризации
        alpha - скорость спуска
        max_epoch - максимальное количество эпох
        """
        
        self.C = C
        self.alpha = alpha
        self.max_epoch = max_epoch
        self.random_state = random_state
    
    def fit(self, X, y=None):
        '''
        Обучение модели
        '''
        #добавим в начало Х столбец с единицами - для свободного члена
        X = np.c_[np.ones(X.shape[0]), X]
        
        #оставляем первый столбец в у, приводим классы к виду +1 или -1
        y[y == 0] = -1
        
        np.random.seed(self.random_state)
        #self.beta = np.random.rand(X.shape[1])
        self.beta = np.zeros(X.shape[1], dtype=np.float64)
        new_beta = np.zeros(X.shape[1], dtype=np.float64)
        self.errors = []
        
        for epoch in xrange(1, self.max_epoch+1):
            perm = np.random.permutation(X.shape[0])
            X = X[perm]
            y = y[perm]
            for it in xrange(X.shape[0]):
                new_beta += (self.alpha/epoch) * self._grad(X[it], y[it], self.beta)
                np.copyto(self.beta, new_beta)  
            self._count_error(X, y, self.beta)
        
        return self
    
    def partial_fit(self, X, y=None):
        '''
        По желанию - метод дообучения модели на новых данных
        '''
    
        return self
        
    def predict(self, X):
        '''
        Возвращение метки класса
        lbl = sign(g(x))
        '''
        #добавим в начало Х столбец с единицами - для свободного члена
        X = np.c_[np.ones(X.shape[0]), X]

        y_hat = np.sign(np.dot(self.beta, X.T))
        y_hat[y_hat == 0] = 1
        return y_hat
    
    def predict_proba(self, X):
        '''
        Возвращение вероятности каждого из классов
        P = sigma(g(x))
        '''
        #добавим в начало Х столбец с единицами - для свободного члена
        X = np.c_[np.ones(X.shape[0]), X]
        
        return 1/(1 + np.exp(np.dot(self.beta, X.T)))
    
    def _grad(self, X, y, beta):
        '''
        Градиент ф-ции потерь, рассчитанный по одной точке 
        ''' 
        #градиент без учета регуляризации
        grad_without_regul = X*y/(1 + np.exp(y*np.dot(X, beta.T)))
        
        #убираем из регуляризации свободный член
        beta[0] = 0
        
        return grad_without_regul - self.C * np.sign(beta)
    
    def _count_error(self, X, y, beta):
        '''
        Вычисляет долю неверно классифицированных объектов
        '''
        lbl = y*np.dot(beta, X.T)
        self.errors.append(float(len(lbl[lbl < 0]))/X.shape[0])
        return
```
